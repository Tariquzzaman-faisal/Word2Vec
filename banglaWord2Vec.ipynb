```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_bn = stopwords.words('bengali')
```

```python
list_[288]
```

```python
new_list = [x for x in list_ if not isinstance(x[0], float)]
len(new_list)
```

```python
sent = new_list
```

```python
tokenized = []
for i in range(len(sent)):
    if i%1000==0:
        print(f'read {i} articles')
    token = nltk.word_tokenize(sent[i][0])
    filtered_words = [word for word in token if (word not in stopwords_bn and word not in [',', '।', ')', '('])]    
    tokenized.append(filtered_words)
```

```python
from bnlp import BengaliWord2Vec
bwv = BengaliWord2Vec()
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
data_file = tokenized
model_name = "test_model.model"
vector_name = "test_vector.vector"
bwv.train(data_file, model_name, vector_name, epochs=5)
```

```python

model_path = "test_model.model"
word = "বিক্ষোভ"
similar = bwv.most_similar(model_path, word, topn=10)
```
